```python
import pandas as pd

df = pd.read_csv("../data/distances_errors_resnet_50.csv")
df
```

```python
from matplotlib import pyplot as plt
import numpy as np

m, b = np.polyfit(df['Distances'], df['Errors'], 1)
ma, ba = np.polyfit(df['Distances'], df['Distances'] * df['Errors'], 1)

# df.plot(x='Distances', y='Errors')
plt.plot(df['Distances'], df['Errors'] * df['Distances'])
# plt.plot(df['Distances'], df['Distances']*m+b)
plt.plot(df['Distances'], df['Distances']*ma+ba)
```
